```python
import os
import glob
import numpy as np
import dask
import dask.array as da
from PIL import Image
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
images = sorted(glob.glob("/home/skyolia/JupyterProjects/segmentation/TGS Salt/dataset/train/images/*.png"))
masks = sorted(glob.glob("/home/skyolia/JupyterProjects/segmentation/TGS Salt/dataset/train/masks/*.png"))
train_idx = np.random.choice(4000, 2400, False)
test_idx = np.nonzero(np.isin(np.arange(4000), train_idx, True, True))[0]
images, masks = np.asarray(images), np.asarray(masks)
images[254], masks[254], train_idx.shape, test_idx.shape, images.shape, masks.shape
```

```python
def load_save(path):
    image = Image.open(path).convert('L')
    image = np.asarray(image)/255.0
    return image

x = []
for i in images:
    image = dask.delayed(load_save)(i)
    x.append(image)
x = np.asarray(dask.compute(*x, scheduler='processes'))

y = []
for i in masks:
    mask = dask.delayed(load_save)(i)
    y.append(mask)
y = np.asarray(dask.compute(*y, scheduler='processes'))
x.shape, y.shape
```

```python
x, y = x.astype(np.float32), y.astype(np.float32)
train_x, train_y = x[train_idx], y[train_idx]
test_x, test_y = x[test_idx], y[test_idx]
train_x.shape, train_y.shape, test_x.shape, test_y.shape
```

```python
train_x = da.from_array(train_x, chunks=(2400, 101, 101))
train_y = da.from_array(train_y, chunks=(2400, 101, 101))
test_x = da.from_array(test_x, chunks=(1600, 101, 101))
test_y = da.from_array(test_y, chunks=(1600, 101, 101))
```
