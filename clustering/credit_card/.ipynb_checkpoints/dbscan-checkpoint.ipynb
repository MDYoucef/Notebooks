```python
from umap import UMAP
import pandas as pd
import numpy as np
import plotly
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score, train_test_split
from catboost import CatBoostClassifier as CBC, Pool
from sklearn.tree import DecisionTreeClassifier as DTC
from sklearn import tree
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn import tree
from dtreeviz.trees import *
import hdbscan
```

```python
#https://www.kaggle.com/sabanasimbutt/clustering-visualization-of-clusters-using-pca
umap = pd.read_csv('/home/skyolia/JupyterProjects/data/credit_card/reduced_data.csv', sep=',')
umap
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
cols = umap.columns
umap = scaler.fit_transform(umap)
umap = pd.DataFrame(data=umap, columns=cols)
umap
```

```python
hp = [200, 70, 0]
clusterer = hdbscan.HDBSCAN(min_cluster_size=int(hp[0]), min_samples=int(hp[1]), cluster_selection_epsilon=float(hp[2]))
cluster_labels = clusterer.fit_predict(umap)
idx = np.where(cluster_labels > -1)[0]
clusterer.cluster_persistence_, np.unique(cluster_labels, return_counts=True), metrics.silhouette_score(umap.iloc[idx, :], cluster_labels[idx], metric='euclidean')
```

```python
fig = go.Figure()
for i in np.unique(cluster_labels):
    idx = np.where(cluster_labels==i)[0]
    fig.add_trace(go.Scatter(x=umap.iloc[idx,0], y=umap.iloc[idx,1], mode='markers', name='cluster_'+str(i)))
fig.show()
```

```python
df = pd.read_csv('/home/skyolia/JupyterProjects/data/credit_card/CC GENERAL.csv', sep=',')
df['MINIMUM_PAYMENTS'].fillna((df['MINIMUM_PAYMENTS'].median()), inplace=True)
df['CREDIT_LIMIT'].fillna((df['CREDIT_LIMIT'].median()), inplace=True)
df['Clusters'] = cluster_labels
df.drop(df[df['Clusters'] == -1].index, inplace=True)
#df.to_csv('/home/skyolia/JupyterProjects/data/credit_card/clustered_data.csv', index=False, sep=',')
```

```python
df = pd.read_csv('/home/skyolia/JupyterProjects/data/credit_card/clustered_data.csv', sep=',')
df
```

```python
df.drop(columns=['CUST_ID'], inplace=True)
df.isna().sum()
```

```python
df['Clusters'].unique()
```

```python
df['RANDOM'] = np.random.uniform(0,100,len(df))
x, y = df[[i for i in df.columns if i!='Clusters']], df[['Clusters']]
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.2, random_state = 0)
columns = list(train_x.columns)
#train_x, test_x, train_y, test_y = train_x.values, test_x.values, train_y.values, test_y.values

hp = [3.80988681e-05, 3.32408151e-05, 1.70000000e+01]
hp = [hp[0], hp[1], np.round(hp[2]).astype(int)]
clf = DTC(min_samples_split=hp[0], min_samples_leaf=hp[1], max_features=hp[2], random_state=22)
clf.fit(train_x, train_y)
clf.score(train_x, train_y), clf.score(test_x, test_y)
```

```python
pred = clf.predict(test_x)
confusion_matrix(test_y, pred), accuracy_score(test_y, pred)
```

```python
#tree.plot_tree(clf, feature_names=train_x.columns, class_names=['cl0', 'cl1'])
viz = dtreeviz(clf, train_x.values, train_y.values.ravel(), target_name='Cluster',feature_names=list(test_x.columns), 
               class_names=['cl_'+str(i) for i in np.unique(train_y)]  # need class_names for classifier
              )
viz.view() 
```

```python
from sklearn.inspection import permutation_importance
from sklearn.metrics import fbeta_score, make_scorer


result = permutation_importance(clf, test_x, test_y, n_repeats=10, random_state=42, n_jobs=2, scoring='f1_weighted')
fig = go.Figure()
# Use x instead of y argument for horizontal plot
for i in range(train_x.shape[1]):
    fig.add_trace(go.Box(x=result['importances'][i], name=columns[i]))

fig.show()
df.drop(columns=['RANDOM'], inplace=True)
columns.remove('RANDOM')
```

```python
train_summary = df.groupby(by='Clusters').mean()
train_summary = train_summary[columns]
train_summary
```

```python
fig = go.Figure()
class_names = ['Cluster_'+str(i) for i in train_summary.index.tolist()]

for col in columns:
    fig.add_trace(go.Scatterpolar(r=train_summary[col].tolist(),theta=class_names,fill='toself',name=col))



fig.show()
```

```python
fig = make_subplots(rows=len(columns), cols=df['Clusters'].nunique(), shared_yaxes=True)
agg = df.groupby(by='Clusters')

for ii, i in enumerate(columns):
    for jj, j in enumerate(agg):
        fig.add_trace(go.Histogram(x=j[1][i].tolist(), name=str(i)+'_'+str(jj)), row=ii+1, col=jj+1)

fig.update_layout(width=4000, height=4000, title_text="Distributions plot hdbscan")
plotly.offline.plot(fig, filename = 'dist_plot_hdbscan.html', auto_open=False)
```

```python
fig = make_subplots(rows=1, cols=len(columns))

for i, col in enumerate(columns):
    fig.add_trace(go.Bar(name=col, x=train_summary.index.tolist(), y=train_summary[col].tolist())
                  , row=1, col=i+1)
    fig.add_trace(go.Scatter(x=train_summary.index.tolist(),y=[df[col].median()]*len(train_summary.index.tolist())
                            , line_color='rgb(0,0,0)', showlegend=False, mode='lines',), row=1, col=i+1)

fig.update_layout(width=4000, title_text="Mean plot by features hdbscan")
plotly.offline.plot(fig, filename = 'mean_plot_by_feat_hdbscan.html', auto_open=False)
#fig.show()
```
