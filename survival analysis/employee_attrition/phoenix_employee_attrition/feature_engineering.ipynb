```python
import pandas as pd
import numpy as np
import zipfile
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```python
in_time = pd.read_csv('/content/drive/My Drive/Colab Notebooks/survival analysis/phoenix_employee_attrition/phoenix_data/in_time.csv')
out_time = pd.read_csv('/content/drive/My Drive/Colab Notebooks/survival analysis/phoenix_employee_attrition/phoenix_data/out_time.csv')
out_time
```

```python
in_time
```

```python
time_df = pd.concat((in_time, out_time))
time_df.rename(columns={'Unnamed: 0': 'EmployeeID'},inplace=True)
time_df = time_df.fillna(0.)
time_df.iloc[:, 1:] =time_df.iloc[:, 1:].apply(pd.to_datetime, errors='coerce')
time_df
```

```python
tmp = time_df.groupby(['EmployeeID']).diff()/np.timedelta64(1, 'h')#.drop_duplicates(subset=["id"], keep="last")
time_df = tmp.reset_index().drop_duplicates(subset='index', keep='last').set_index('index')
time_df['Avg'] = time_df.median(axis=1)
time_df
```

```python
time_df = time_df[['Avg']]
time_df['EmployeeID'] = time_df.index + 1
time_df.reset_index(drop=True, inplace=True)
time_df
```

```python
general_df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/survival analysis/phoenix_employee_attrition/phoenix_data/general_data.csv')
employee_df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/survival analysis/phoenix_employee_attrition/phoenix_data/employee_survey_data.csv')
manager_df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/survival analysis/phoenix_employee_attrition/phoenix_data/manager_survey_data.csv')
general_df
```

```python
manager_df
```

```python
employee_df
```

```python
df = pd.merge(employee_df, manager_df, how='inner', on='EmployeeID')
df = pd.merge(general_df, df, how='inner', on='EmployeeID')
df = pd.merge(time_df, df, how='inner', on='EmployeeID')
df
```

```python
df.isna().sum()
```

```python
df['EnvironmentSatisfaction'].fillna(df['EnvironmentSatisfaction'].median(),inplace = True)
df['JobSatisfaction'].fillna(df['JobSatisfaction'].median(),inplace = True)
df['WorkLifeBalance'].fillna(df['WorkLifeBalance'].median(),inplace = True)
df['NumCompaniesWorked'].fillna(df['NumCompaniesWorked'].median(),inplace = True)
df['TotalWorkingYears'].fillna(df['TotalWorkingYears'].median(),inplace = True)
```

```python
df.drop(columns=['Over18', 'EmployeeID', 'EmployeeCount', 'StandardHours'], inplace=True, errors='ignore')
```
